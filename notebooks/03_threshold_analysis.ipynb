```python

{
 "cells": [
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "# 03 - Análise de Threshold\n",
    "\n",
    "Este notebook avalia diferentes pontos de corte para equilibrar Recall e FPR."
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import pandas as pd\n",
    "import numpy as np\n",
    "from sklearn.metrics import recall_score, confusion_matrix\n",
    "import matplotlib.pyplot as plt\n",
    "\n",
    "df = pd.read_parquet('../data/processed/dataset.parquet')\n",
    "from joblib import load\n",
    "model = load('../models/artifacts/model_xgb.joblib')\n",
    "\n",
    "X = df.drop(columns=['is_fraud'])\n",
    "y = df['is_fraud']\n",
    "proba = model.predict_proba(X)[:,1]\n",
    "\n",
    "thresholds = np.linspace(0.1, 0.99, 20)\n",
    "recalls, fprs = [], []\n",
    "for t in thresholds:\n",
    "    y_pred = (proba >= t).astype(int)\n",
    "    recalls.append(recall_score(y, y_pred))\n",
    "    cm = confusion_matrix(y, y_pred)\n",
    "    fp = cm[0,1]\n",
    "    tn = cm[0,0]\n",
    "    fprs.append(fp / (fp+tn))\n",
    "\n",
    "plt.plot(thresholds, recalls, label='Recall')\n",
    "plt.plot(thresholds, fprs, label='FPR')\n",
    "plt.xlabel('Threshold')\n",
    "plt.ylabel('Métrica')\n",
    "plt.legend()\n",
    "plt.title('Trade-off Recall vs FPR')\n",
    "plt.show()"
   ]
  }
 ],
 "metadata": {
  "kernelspec": {
   "display_name": "Python 3",
   "language": "python",
   "name": "python3"
  }
 },
 "nbformat": 4,
 "nbformat_minor": 2
}
```
